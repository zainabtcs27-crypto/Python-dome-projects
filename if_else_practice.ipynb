The Grade Calculator (Basic) Ek program likhein jo user se marks (0-100) le aur niche diye gaye criteria ke mutabiq grade print kare:
90 aur usse upar: Grade A

80 se 89: Grade B

70 se 79: Grade C

60 se niche: Fail

```python
Marks = int(input("Enter your marks: "))

if Marks >= 90:
    print("Grade A")
elif Marks >= 80 and Marks <= 89:
    print("Grade B")
elif Marks >= 70 and Marks <= 79:
    print("Grade C")
else:
    print("Fail")
```

2. Login System (Logical Operators)

3. Leap Year Checker (Mathematical Logic)
User se ek saal (year) input lein aur check karein ke wo Leap Year hai ya nahi.

Hint: Leap year wo hota hai jo 4 se divide ho jaye, lekin agar wo 100 se divide hota hai to use 400 se bhi divide hona chahiye.

```python
Year = int(input("Enter your year: "))
if Year % 400 == 0  :
    print("Leap Year")
elif Year % 100 == 0 :
    print("Not a Leap Year")
elif Year % 4 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

5. ATM Simulation (Nested If-Else)
Ek variable rakhein balance = 5000. User se do cheezein puchein:

Wo kya karna chahte hain? (Withdraw ya Deposit)

Amount kitni hai?

Agar Withdraw select karein aur amount balance se zyada ho, to kahein "Insufficient Funds".

Agar amount balance se kam ho, to balance minus karein aur naya balance dikhayein.

Agar Deposit ho, to balance mein add karein.

```python
balance = 5000
user_desire = str(input("Enter your desire: "))
if user_desire == "withdraw":
    amount = int(input("Enter your amount to withdraw: "))
    if amount<=balance:
        balance = balance - amount
        print("Your remaining balance is: ", balance)
    else:
        print("Insufficient balance")
elif user_desire == "deposit":
    amount = int(input("Enter your amount to deposit: "))
    balance=amount+balance
    print("Your amount after deposit:", balance)
else:
    print("Invalid Desire")
```

1. Simple Calculator (Multiple Elif)
User se do numbers lein aur phir ek operator input karwayein (jaise +, -, *, /).

```python
num1 = int(input("Enter your first number: "))
num2 = int(input("Enter your second number: "))
operator = str(input("Enter your operator: "))
if  operator == "+" :
    result = num1 + num2
    print(num1,operator,num2,"=",result)
elif operator == "-" :
    result = num1 - num2
    print(num1,operator,num2,"=",result)
elif operator == "*" :
    result = num1 * num2
    print(num1,operator,num2,"=",result)
elif operator == "/" :
    if num2 == 0:
        print("You cannot divide by zero")
    else:
        result = num1 / num2
        print(num1,operator,num2,"=",result)
else:
    print("Invalid operator")
```

2. Movie Ticket Pricing (Age & Day Logic)
Ek cinema ka system banayein:

Agar age 12 se kam ho, to ticket 500 hai.

Agar age 12 se 60 ke darmiyan ho, to ticket 1000 hai.

Agar age 60 se upar ho, to ticket 700 hai (Senior citizen discount).

Bonus: Agar aaj "Wednesday" hai, to har ticket par mazeed 20% discount dein.

```python
import string

print(dir(string))
```

```python
age = int(input("Enter your age: "))
day = str(input("Whats the day today: ")).lower()
if age < 12:
    if day == "Wednesday".lower():
        discounted_amount = (500*20)/100
        discounted_price = 500 - discounted_amount
        print("Discounted price on 'Wednesday' for under 12 years is: ", discounted_price)
    else:
        print("Ticket price for under 12: 500")
elif age >= 12 and age <=60:
    if day == "Wednesday".lower():
        discounted_amount = (1000*20)/100
        discounted_price = 1000 - discounted_amount
        print("Discounted price on 'Wednesday' from 12 to 60 years is: ", discounted_price)
    else:
        print("Ticket price from 12 to 60 years: 1000")
elif age >60:
    if day == "Wednesday".lower():
        discounted_amount = (700*20)/100
        discounted_price = 700 - discounted_amount
        print("Discounted price on 'Wednesday' up to 60 years is: ", discounted_price)
    else:
        print("Ticket price up to 60 years: 700")
else:
    print("Invalid")
```
